# Multi-layer perceptron

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten,Dense
```

```python
(X_train,y_train),(X_test,y_test)=tf.keras.datasets.mnist.load_data()
```

```python
gray_scale=225
X_train=X_train.astype("float32")/gray_scale
X_test=X_test.astype("float32")/gray_scale
```

```python
print("shape of x_train data is",X_train.shape)
print("shape of X_test data is",X_test.shape)
print("shape of y__train data is",y_train.shape)
print("shape of y_test data is ",y_test.shape)
```

```python
fig,ax=plt.subplots(10,10)
k=0
for i in range(10):
    for j in range(10):
        ax[i][j].imshow(X_train[k].reshape(28,28),aspect="auto")
        k+=1
plt.show()
```

```python
# creating an model with the the sequential model and dense layers
model=Sequential([
    Flatten(input_shape=(28,28)),
    Dense(256,activation="sigmoid"),
    Dense(125,activation="sigmoid"),
    Dense(10,activation="softmax")])
```

```python
model.compile(optimizer="adam",
              loss="sparse_categorical_crossentropy",
              metrics=(["accuracy"])
             )
```

```python
model.fit(X_train,y_train,epochs=10,batch_size=2000,validation_split=0.2)
```

```python
model.evaluate(X_test,y_test,verbose=10)
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Flatten,Dense
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import SparseCategoricalCrossentropy
```

```python
from tensorflow.keras.metrics import SparseCategoricalAccuracy
```

```python
mnist=tf.keras.datasets.mnist
```

```python
(X_train,y_train),(X_test,y_test)=mnist.load_data()
```

```python
model=Sequential([
    Flatten(input_shape=(28,28)),
    Dense(125,activation="relu"),
    Dense(10,activation="softmax")])
```

```python
model.compile(optimizer=Adam(),
              loss=SparseCategoricalCrossentropy(),
              metrics=[SparseCategoricalAccuracy])
              
```

```python
model.fit(X_train,y_train,epochs=5)
```

```python
loss,accuracy=model.evaluate(X_test,y_test)
```

```python
print(f"accuracy is: {accuracy},loss is {loss}")
```
