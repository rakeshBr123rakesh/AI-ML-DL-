#### Develop a program to demonstrate the working of Linear Regression and Polynomial Regression. Use Boston Housing Dataset for Linear Regression and Auto MPG Dataset (for vehicle fuel efficiency prediction) for Polynomial Regression.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression 
from sklearn.datasets import fetch_openml
```

```python
# Load boaster Housing dataset from OpenML
boston=fetch_openml(name="boston",version=1,as_frame=True)
x_boston=boston.data
y_boston=boston.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    x_boston[['LSTAT']], y_boston, test_size=0.2, random_state=42
)
```

```python
lin_reg=LinearRegression()
lin_reg.fit(X_train,y_train)
y_pred=lin_reg.predict(X_test)
```

```python
# Plot results
plt.scatter(X_test, y_test, color="blue", label="Actual Data", alpha=0.5)
plt.plot(X_test, y_pred, color="red", linewidth=2, label="Linear Regression")
plt.xlabel("% Lower Status of Population (LSTAT)")
plt.ylabel("House Price (in $1000s)")
plt.legend()
plt.title("Linear Regression on Boston Housing Dataset")
plt.show()
```
